<h4><center>**The commands that we used in this project :</center></h4

<h5>
1.head() - It shows the first N rows in the data (by default, N=5).<br>
2.tail () - It shows the last N rows in the data (by default, N=5).<br>
3.shape - It shows the total no. of rows and no. of columns of the dataframe.<br>
4.size - To show No. of total values(elements) in the dataset.<br>
5.columns - To show each Column Name.<br>
6.dtypes - To show the data-type of each column.<br>
7.info() - To show indexes, columns, data-types of each column, memory at once.<br>
8.value_counts - In a column, it shows all the unique values with their count. It can be applied on a single column only.<br>
9.unique() - It shows the all unique values of the series.<br>
10.nunique() - It shows the total no. of unique values in the series.<br>
11.duplicated( ) - To check row wise and detect the Duplicate rows.<br>
12.isnull( ) - To show where Null value is present.<br>
13.dropna( ) - It drops the rows that contains all missing values.<br>
14.isin( ) - To show all records including particular elements.<br>
15.str.contains( ) - To get all records that contains a given string.<br>
16.str.split( ) - It splits a column's string into different columns.<br>
17.to_datetime( ) - Converts the data-type of Date-Time Column into datetime[ns] datatype.<br>
18.dt.year.value_counts( ) - It counts the occurrence of all individual years in Time column.<br>
19.groupby( ) - Groupby is used to split the data into groups based on some criteria.<br>
20.sns.countplot(df['Col_name']) - To show the count of all unique values of any column in the form of bar graph.<br>
21.max( ), min( ) - It shows the maximum/minimum value of the series.<br>
22.mean( ) - It shows the mean value of the series.<br>
